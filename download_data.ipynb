```python
import os
import requests
import zipfile
import io

# Configura√ß√µes
DATASET_URL = "https://storage.googleapis.com/magentadata/datasets/maestro/v3.0.0/maestro-v3.0.0-midi.zip"
DEST_FOLDER = "data"

def download_and_extract():
    print(f"üöÄ Iniciando download do dataset MAESTRO ({DATASET_URL})...")
    print("‚òï Isso pode levar alguns minutos dependendo da sua internet...")
    
    try:
        # Faz o download
        response = requests.get(DATASET_URL, stream=True)
        response.raise_for_status()
        
        # Cria a pasta data se n√£o existir
        if not os.path.exists(DEST_FOLDER):
            os.makedirs(DEST_FOLDER)
            
        print("üì¶ Extraindo arquivos...")
        
        # Descompacta em mem√≥ria (sem salvar o zip no disco para economizar espa√ßo)
        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            z.extractall(DEST_FOLDER)
            
        print(f"‚úÖ Sucesso! Dataset pronto em: {os.path.abspath(DEST_FOLDER)}")
        print("üéπ Agora voc√™ pode rodar os notebooks!")
        
    except Exception as e:
        print(f"‚ùå Erro: {e}")

if __name__ == "__main__":
    download_and_extract()
```
